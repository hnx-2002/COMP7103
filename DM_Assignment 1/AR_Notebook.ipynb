# Question 1:
<p style="font-size:22px;">Report 3 rules with support at least 0.2 and confidence at least 0.9. Specify for each of them the support and the confidence.</p>

```python
import math
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules\


data = pd.read_csv(r"mammographic_masses.csv",delimiter=",",header=0)

#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()


# NAN == "?"
for i in range(len(d)):
    j = 0
    while True:
        if d[i][j] == '?':  
            del d[i][j]     
            j -= 1          
        j += 1
        if j > len(d[i]) - 1:
            break


#adding attributes
for i in range(len(d)):
    for j in range (len(d[i])):
        d[i][j]=data.columns[j] + "=" +str(d[i][j])

            
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)

df = pd.DataFrame(te_ary, columns=te.columns_)

#computing frequent itemsets and association rules
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

a=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

selected_rules = a[(a['support'] >= 0.2) & (a['confidence'] >= 0.9)]
selected_rules[["antecedents", "consequents", "support", "confidence"]]
```

<p style="font-size:20px;">So, The table displays all rules with support >= 0.2 and confidence >= to 0.9. And the the rules along with their actual support and confidence are displayed.</p>

# Question 2
<p style="font-size:22px;">Mining the rules that can help us to predicting whether a given instance is benign or malign.</p>

```python
import math
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules\


data = pd.read_csv(r"mammographic_masses.csv",delimiter=",",header=0)

#preprocessing: change to one hot encoding so as to be able to use apriori from mlxtend
d=data.values.tolist()


# NAN == "?"
for i in range(len(d)):
    j = 0
    while True:
        if d[i][j] == '?':  
            del d[i][j]     
            j -= 1          
        j += 1
        if j > len(d[i]) - 1:
            break


#adding attributes
for i in range(len(d)):
    for j in range (len(d[i])):
        d[i][j]=data.columns[j] + "=" +str(d[i][j])

            
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)

df = pd.DataFrame(te_ary, columns=te.columns_)

#computing frequent itemsets and association rules
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

rule = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.9)

rule[["antecedents", "consequents", "support", "confidence"]]

```

```python
# exclude BI-RADS and predict the severity

severity_rule = rule[
    (rule['support'] >= 0.1) &
    (rule['consequents'].apply(lambda x: 'Severity=0' in x or 'Severity=1' in x)) &
    (~rule['antecedents'].apply(lambda x: any('BI-RADS' in item for item in x)))
]

severity_rule[["antecedents", "consequents", "support", "confidence"]]

```

<p style="font-size:20px;">When the Density of the lesion is 3, the Shape of it is round and the Margin of it is circumscribed, then the lesions are usually benign.
This rule has support of 14.36% in the dataset without BI-RADS. And it is correct in 90.20% of cases (90.20% Confidence). </p>

# Question 3
<p style="font-size:22px;">Find some rules proving that BI-RADS is not always precisely indicate benign or malign</p>

<p style="font-size:20px;">Reset the confidence threshold to 0.7. It is to ensure that the rules found have a high level of confidence, while increasing the number of potentially useful rules found for subsequent screening. This approach can explore the relationship between BI-RADS and Severity more extensively.</p>

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Generate association rules
rules_birads = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Filter rules where 'BI-RADS' is part of the antecedents or consequents and relates to 'Severity'
birads_inaccuracy_rules = rules_birads[
    (rules_birads['antecedents'].apply(lambda x: any('BI-RADS' in item for item in x)) |
     rules_birads['consequents'].apply(lambda x: any('BI-RADS' in item for item in x))) &
    (rules_birads['consequents'].apply(lambda x: 'Severity=0' in x or 'Severity=1' in x))
]

birads_inaccuracy_rules[["antecedents", "consequents", "support", "confidence"]]

```

<p style="font-size:20px;">We have discovered many rules here, including BI-RADS and Severity, but still need to refine.</p>

```python
birads_inaccuracy_rules2 = rules_birads[
    (rules_birads['antecedents'].apply(lambda x: any('BI-RADS' in item for item in x))) &
    (rules_birads['consequents'].apply(lambda x: 'Severity=1' in x))
]

birads_inaccuracy_rules2[["antecedents", "consequents", "support", "confidence"]]
```

<p style="font-size:20px;">Here, we find that the antecedents containing BI-RADS that can derive Severity=1 are all "BI-RADS=5". The high value of BI-RADS indicates a high possibility of malignancy, which is consistent with the consequents terms. So need to mining deeper.</p>

```python
birads_inaccuracy_rules3 = rules_birads[
    (rules_birads['antecedents'].apply(lambda x: any('BI-RADS' in item for item in x))) &
    (rules_birads['consequents'].apply(lambda x: 'Severity=0' in x))
]

birads_inaccuracy_rules3[["antecedents", "consequents", "support", "confidence"]]
```

<p style="font-size:20px;">Here, we search for the rules that have BI-RADS in its antecedents and Severity=0 in its consequents. Then we find something interesting.</p>

```python

specific_rule = rules_birads[
    (rules_birads['antecedents'].apply(lambda x: 'BI-RADS=4' in x)) &
    (rules_birads['consequents'].apply(lambda x: 'Severity=0' in x))
]


specific_rule[["antecedents", "consequents", "support", "confidence"]]
```

<p style="font-size:20px;">
So, We have a new finding. When BI-RADS=4, which means a slightly higher possibility of malignancy, the Severity is 0. It contradicts the description of BI-RADS. Normally, when BI-RADS has a high value, Severity tends to be more malignant.<br>
Even the confidence of the rule (Density=3, Margin=1, BI-RADS=4) ->(Severity=0) has reached 90%.
</p>

# Question 4
<p style="font-size:22px;">Find the rules and its support and confidence: Age=35 -> Severity=0.</p>

```python
antecedent = df['Age=35']
consequent = df['Severity=0']

support = ((antecedent & consequent).sum()) / len(df)

confidence = ((antecedent & consequent).sum()) / antecedent.sum()

print(f"Support: {support:.6f}, Confidence: {confidence:.6f}")
```

<p style="font-size:20px;">
We find the rules: Age=35 -> Severity=0.<br><br>
It's Support: 0.010406.<br>
It's Confidence: 0.769231.<br><br>
This meas even though the confidence indicates that in the case of Age=35, most instances are benign, there are only 1.04% of the instances in the entire dataset meet the criteria of Age=35 and Severity=0. This value is really really low, indicating that this rule appears very infrequently in the dataset.<br><br>

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

n = 60

data_missing['Age_binary'] = data_missing['Age'].apply(lambda x: 1 if x >= n else 0)


data_list = data_missing.values.tolist()

# dealing with tht missing values exclude 'Age'
for i in range(len(data_list)):
    j = 0
    while j < len(data_list[i]):
        # is NAN?
        if pd.isna(data_list[i][j]) and data_missing.columns[j] != 'Age':  
            del data_list[i][j]  # del
            j -= 1
        j += 1

# add clo name to mining the rules
for i in range(len(data_list)):
    for j in range(len(data_list[i])):
        data_list[i][j] = data_missing.columns[j] + "=" + str(data_list[i][j])


te_age = TransactionEncoder()
te_array_age = te_age.fit(data_list).transform(data_list)
df_age = pd.DataFrame(te_array_age, columns=te_age.columns_)


freq_itemsets = apriori(df_age, min_support=0.1, use_colnames=True)
ar_ages = association_rules(freq_itemsets, metric="confidence", min_threshold=0.9)


ar_ages[["antecedents", "consequents", "support", "confidence"]]
```

```python
final_rules = ar_ages[
    ar_ages['antecedents'].apply(lambda x: any('Age_binary=1' in item for item in x)) &
    ar_ages['consequents'].apply(lambda x: any('Severity=1' in item for item in x))
]

final_rules[['antecedents', 'consequents', 'support', 'confidence']]
```
